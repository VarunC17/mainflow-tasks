```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\Varun Contractor\\Documents\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#Scatter plot with year against own
plt.scatter(data['year'],data['own'])

#Adding title to the plot
plt.title("Scatter Plot")

#Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#showing the result
plt.show()
```

```python
#Line chart(year against own)
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("Line chart")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#Bar chart(year against expenditure)
plt.bar(data['year'],data['expenditure'])
plt.title("Bar Chart")
plt.xlabel('year')
plt.ylabel('expenditure')
plt.show()
```
